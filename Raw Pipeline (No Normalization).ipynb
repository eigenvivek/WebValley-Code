```python
import os
import numpy as np
from PIL import Image
import sqlite3 as lite
import matplotlib.pyplot as plt
import pandas as pd
```

```python
full_dir = './Data'
dates = os.listdir(full_dir)
```

```python
print(dates)
```

```python
for date in dates:
    data_dir = './Data/' + date + '/images/'
    img_dir = os.path.join(data_dir)
    out_dir = os.path.join(data_dir, 'cropped')
    spect_dir = os.path.join(data_dir, 'spectra')
    
    try:
        os.mkdir(out_dir)
        os.mkdir(spect_dir)
    except OSError:
        pass
    
    # Connect to DB
    to_execute = 'SELECT * FROM samples;'
    con = lite.connect(os.path.join(data_dir, 'samples.db'))
    cur = con.cursor()
    cur.execute(to_execute)
    con.commit()
    results = cur.fetchall()
    con.close()
    
    bbox = [800, 1000, 3000, 1300]
    for r in results:
        print('Cropping image: {} from {}'.format(r[-2].lower(), r[-1].lower()))
        # Get the image
        im = Image.open(os.path.join(img_dir, r[-2]))
        
        # Crop the image with the interesting part
        region = im.crop(bbox)

        # Save the region
        region.save(os.path.join(out_dir, '{}.jpg'.format(r[-1].lower())))
        regarr = np.array(region)

        # Compute spectrum from the image
        spectrum = regarr.sum(axis=2).mean(axis=0)
        
        np.savetxt(os.path.join(spect_dir, '{}.csv'.format(r[-1].lower())),
               spectrum, delimiter='\t')
        
    # Concatenate the CSVs
    
    parent = str('/home/wvstudent/notebooks/Spectroscopy/data/' +date+ '/images/spectra')
    
    files = os.listdir(parent)

    A = len(files)
    N = 1

    merged = []

    for f in files:
        print('Merging file ' + str(N) + " of " + str(A))
        read = pd.read_csv(os.path.join(parent, f), header=None)
        read = np.transpose(read)
        merged.append(read)
        N += 1
    
    result = pd.concat(merged)
    result.to_csv(date + '.csv', ',')

    print('DONe!')
```
